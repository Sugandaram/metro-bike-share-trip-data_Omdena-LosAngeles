## I split date & time (day, month, year, hour, minute)

```python
MB3["Start Time"] = pd.to_datetime(MB3["Start Time"])
MB3["Start Time"]
```

```python

MB3["Start year"] = MB3["Start Time"].dt.year
MB3["Start Month"] = MB3["Start Time"].dt.month
MB3["Start Day"] = MB3["Start Time"].dt.day
```

```python
MB3["End Time"] = pd.to_datetime(MB3["End Time"])
MB3["End Time"]
```

```python
MB3["End year"] = MB3["End Time"].dt.year
MB3["End Month"] = MB3["End Time"].dt.month
MB3["End Day"] = MB3["End Time"].dt.day
```

## Directive

```python
MB3["Start Time"] = pd.to_datetime(MB3["Start Time"])
```

```python
MB3["Start Hour"] = MB3["Start Time"].dt.strftime("%H")
MB3["Start Min"] = MB3["Start Time"].dt.strftime("%M")
MB3["Start Sec"] = MB3["Start Time"].dt.strftime("%S")
```

```python
MB3["End Time"] = pd.to_datetime(MB3["End Time"])
```

```python
MB3["End Hour"] = MB3["End Time"].dt.strftime("%H")
MB3["End Min"] = MB3["End Time"].dt.strftime("%M")
MB3["End Sec"] = MB3["End Time"].dt.strftime("%S")
```

```python
MB3.columns
```

```python
MB3.drop(['Start Time', 'End Time'],axis = 1, inplace = True)
```

```python
MB3.head(2)
```

```python
MB3["Start Hour"] = MB3["Start Hour"].astype('int')
MB3["Start Min"] = MB3["Start Min"].astype('int')
MB3["Start Sec"] = MB3["Start Sec"].astype('int')
MB3["End Hour"] = MB3["End Hour"].astype('int')
MB3["End Min"] = MB3["End Min"].astype('int')
MB3["End Sec"] = MB3["End Sec"].astype('int')
```
